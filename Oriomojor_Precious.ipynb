```python
#converting order date to date time for anlysis purpose 
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%Y', errors='coerce')
```

```python
# creating a column for YearMonth 
df['YearMonth'] = df['Order Date'].dt.to_period('M').astype(str)
```

```python
df['YearMonth']
```

## Part 1.2: Data Aggregation & Summarisation


```python
# Table A: Monthly Profit & Sales
TableA = df.groupby('YearMonth').agg(
    Total_Sales=('Sales', 'sum'),
    Total_Profit=('Profit', 'sum')
).reset_index()
TableA['YearMonth_dt'] = pd.to_datetime(tableA['YearMonth'])
TableA = tableA.sort_values('YearMonth_dt').reset_index(drop=True)
TableA
```

```python
# Table B: Product Performance by Sub-Category 
TableB = df.groupby(['Category', 'Sub-Category']).agg( Total_sales = ('Sales', 'sum'), 
                                                      Total_profit = ('Profit', 'sum'), 
                                                      Average_Discount = ('Discount', 'mean')).reset_index() 
TableB
TableB['Profit_Margin'] = TableB['Total_profit'] / TableB['Total_sales']
TableB.sort_values('Profit_Margin', ascending = False, inplace = True) 
TableB
```

```python
# Table C: Regional Comparison 
TableC = df.groupby('Region').agg(Total_profit = ('Profit', 'sum'), 
                                  Number_of_orders = ('Order ID', pd.Series.nunique)).reset_index() 
TableC
```

```python
TableC.sort_values('Total_profit', ascending = False)
```

## Part 2: Task A — Scatter Plot: Profitability vs. Discount Strategy

```python
import seaborn as sns
```

```python
sns.scatterplot(data=TableB, x='Average_Discount', y='Profit_Margin') 
plt.title('Scatter Plot: Profitability vs. Discount Strategy')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(9, 6))
colors = {'Furniture': '#e74c3c', 'Office Supplies': '#3498db', 'Technology': '#2ecc71'}

for cat in TableB['Category'].unique():
    sub = TableB[TableB['Category'] == cat]
    sizes = (sub['Total_sales'] / TableB['Total_sales'].max()) * 2000 + 50
    ax.scatter(sub['Average_Discount'], sub['Profit_Margin'], s=sizes, c=colors[cat],
               alpha=0.65, edgecolors='black', linewidth=0.7, label=cat)

for _, row in TableB.iterrows():
    ax.annotate(row['Sub-Category'], (row['Average_Discount'], row['Profit_Margin']),
                fontsize=7.5, xytext=(5, 3), textcoords='offset points')

ax.axhline(0, color='gray', linewidth=0.8, linestyle='--')
ax.set_xlabel('Average Discount')
ax.set_ylabel('Profit Margin (Total Profit / Total Sales)')
ax.set_title('Profitability vs. Discount Strategy by Product Sub-Category\n(bubble size = Total Sales)')
ax.legend(title='Category', loc='upper right')
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:.0%}'))
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.0%}'))
plt.tight_layout()
plt.show()

print('Correlation between Average Discount and Profit Margin:', TableB['Average_Discount'].corr(TableB['Profit_Margin']))
```

# There is a strong negative relationship between average discount and profit margin across
# sub-categories (correlation coefficient ≈ −0.88): the more heavily a sub-category is discounted, the lower—or more negative—its profit margin.
# High-margin sub-categories such as Paper (24.2% margin), Labels (20.4%), and Copiers (17.1%)all carry comparatively modest average discounts (11–12%), reinforcing that disciplined discountingprotects profitability.

## Part 2: Task B — Time Series Plot: Sales & Profit Over Time

```python
fig, ax1 = plt.subplots(figsize=(11, 5.5))
ax1.plot(TableA['YearMonth_dt'], TableA['Total_Sales'], color='#2980b9', linewidth=2, label='Total Sales')
ax1.set_xlabel('Month')
ax1.set_ylabel('Total Sales ($)', color='#2980b9')
ax1.tick_params(axis='y', labelcolor='#2980b9')
ax1.xaxis.set_major_locator(mdates.MonthLocator(interval=3))
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%b-%Y'))
plt.setp(ax1.xaxis.get_majorticklabels(), rotation=45, ha='right')

ax2 = ax1.twinx()
ax2.plot(TableA['YearMonth_dt'], tableA['Total_Profit'], color='#e67e22', linewidth=2, label='Total Profit')
ax2.set_ylabel('Total Profit ($)', color='#e67e22')
ax2.tick_params(axis='y', labelcolor='#e67e22')

lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')
ax1.set_title('Monthly Sales and Profit Trend (2011-2014)')
ax1.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

#Both sales and profit show a consistent upward trend across the four years: annual sales grew from about 2.26M  in 2011 to 4.30M in 2014,

#while annual profit grew from about 249,000, to ,   504,000 over the same period.

#Clear seasonality is visible every year: sales dip in the January–April window and build steadily
#from mid-year, peaking sharply in the September–December period. The single highest month in the dataset is November 2014 ($555,000 in sales)

## Part 2: Task C — Bar Chart: Regional Performance Comparison 

```python
fig, ax1 = plt.subplots(figsize=(11, 6))
x = np.arange(len(TableC))
width = 0.38

bars1 = ax1.bar(x - width/2, TableC['Total_profit'], width, color='#8e44ad', label='Total Profit ($)')
ax1.set_ylabel('Total Profit ($)', color='#8e44ad')
ax1.tick_params(axis='y', labelcolor='#8e44ad')
ax1.set_xticks(x)
ax1.set_xticklabels(TableC['Region'], rotation=45, ha='right')

ax2 = ax1.twinx()
bars2 = ax2.bar(x + width/2, TableC['Number_of_orders'], width, color='#f1c40f', label='Number of Orders')
ax2.set_ylabel('Number of Orders', color='#f1c40f')
ax2.tick_params(axis='y', labelcolor='#f1c40f')

ax1.set_title('Regional Performance: Total Profit vs. Number of Orders')
ax1.set_xlabel('Region')
ax1.legend([bars1, bars2], ['Total Profit ($)', 'Number of Orders'], loc='upper right')
ax1.grid(alpha=0.3, axis='y')
plt.tight_layout()
plt.show()

print('Correlation between Number of Orders and Total_Profit:', TableC['Number_of_orders'].corr(TableC['Total_profit']))
```

#Central is the most profitable region by a wide margin, generating about $311,000 in profit from 5,249 orders—both the highest profit and the highest order count of any region.
